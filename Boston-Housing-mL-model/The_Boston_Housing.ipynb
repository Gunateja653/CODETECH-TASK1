```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/boston.csv")
```

```python
for i in data.columns:
  plt.figure()
  plt.subplot(1,2,1)
  sns.distplot(data[i])
  plt.subplot(1,2,2)
  sns.boxplot(data[i])
  plt.show( )
```

```python
for k,v in data.items():  #k=keys(columns names) ,v=values(columns values)
  q1=v.quantile(0.25)
  q3=v.quantile(0.75)
  iqr=q3-q1
  v_col = v[(v <= q1 - 1.5 * iqr) | (v >= q3 + 1.5 * iqr)]
  perc = np.shape(v_col)[0] * 100.0 / np.shape(data)[0]
  print("Column %s outliers = %.2f%%" % (k, perc))
```

```python
data=data[~(data['MEDV']>=50.0)]
```

```python
for i in data.columns:
  plt.figure(figsize=(12,4))
  sns.scatterplot(data=data,x=i,y='MEDV',hue='CHAS',palette="rainbow")
  plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
x=data[['CRIM','INDUS', 'NOX', 'RM', 'TAX', 'LSTAT']]
y=data['MEDV']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

def selection(model):
    model.fit(x_train,y_train)
    print(f"rmse:{np.sqrt(mean_squared_error(y_test,model.predict(x_test)))}")
    print(f"r2_score:{r2_score(y_test,model.predict(x_test))}")

```

```python
from sklearn.linear_model import LinearRegression
```

```python
selection(LinearRegression())
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
selection(RandomForestRegressor())
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
selection(AdaBoostRegressor())
```

```python
from sklearn.svm import SVR
```

```python
selection(SVR())
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
selection(DecisionTreeRegressor())
```

```python
from xgboost import XGBRegressor
```

```python
selection(XGBRegressor()) #it has the best rmse value than Randomforestreegressor
```
